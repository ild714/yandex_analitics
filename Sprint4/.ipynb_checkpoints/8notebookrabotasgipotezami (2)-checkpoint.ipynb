```python
# этот код служает для спойлеров, выполни его но не читай :)
from IPython.display import display, HTML

def create_spoiler(title, content):
    return HTML(f"""
    <details>
        <summary>{title}</summary>
        <p>{content}</p>
    </details>
    """)

# Пример использования
title = "Нажмите здесь, чтобы увидеть формулировку гипотез"
content_ttest_ind = """
<p>Формулировка гипотез:</p>
<p>H0 (Нулевая гипотеза): Новая рекламная кампания не влияет на продажи (средние значения продаж в двух группах равны)</p>
<p>H1 (Альтернативная гипотеза): Новая рекламная кампания изменяет продажи (средние значения продаж в двух группах различаются)</p>
"""
content_ttest_rel = """
<p>Формулировка гипотез:</p>
<p>H0 (Нулевая гипотеза): Новая методика обучения не влияет на производительность (средние значения производительности до и после равны)</p>
<p>H1 (Альтернативная гипотеза): Новая методика обучения увеличивает производительность (средние значения производительности до и после различаются)</p>
"""
content_ttest_1samp= """
<p>Формулировка гипотез:</p>
<p>H0 (Нулевая гипотеза): Средний уровень удовлетворенности клиентов равен 8</p>
<p>H1 (Альтернативная гипотеза): Средний уровень удовлетворенности клиентов отличается от 8</p>
"""
```

# Работа с гипотезами

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")  # бережем глаза от ненужных предупреждений
```

## Независимый t-тест (ttest_ind)

Бизнес-задача  
Компания хочет проверить, влияет ли новая рекламная кампания на продажи продукта.   
Для этого сравниваются продажи в двух независимых группах: одна группа видела новую рекламу, а другая - нет.

Какие будут гипотезы?

```python
display(create_spoiler(title, content_ttest_ind))
```

```python
# Данные для анализа
np.random.seed(42)

# Данные для анализа
# Генерация данных для группы без рекламы
# loc=50: Среднее значение продаж в группе без рекламы
# scale=10: Стандартное отклонение продаж в группе без рекламы
# size=100: Количество наблюдений в группе без рекламы
sales_no_ad = np.random.normal(loc=50, scale=10, size=100)

# Генерация данных для группы с рекламой
# loc=60: Среднее значение продаж в группе с рекламой
# scale=10: Стандартное отклонение продаж в группе с рекламой
# size=100: Количество наблюдений в группе с рекламой
sales_with_ad = np.random.normal(loc=60, scale=10, size=100)
```

```python
sales_no_ad
```

```python
sales_no_ad.mean()
```

```python
sales_with_ad.mean()
```

```python
# Выполнение t-теста для сравнения средних значений двух независимых групп
t_stat, p_value = st.ttest_ind(sales_no_ad, sales_with_ad)

print('p-value', p_value)

# Оценка результатов
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: новая рекламная кампания изменяет продажи")
else:
    print("Не отвергаем нулевую гипотезу: недостаточно доказательств, что новая рекламная кампания изменяет продажи")

print(f'Среднее без рекламы: {np.mean(sales_no_ad):.2f}')
print(f'Среднее с рекламой: {np.mean(sales_with_ad):.2f}')
```

```python
# Выполнение t-теста для сравнения средних значений двух независимых групп
t_stat, p_value = st.ttest_ind(sales_with_ad, sales_no_ad) # коммент alternative

print('p-value', p_value)
print('t_stat', t_stat)

p_value_one_sided = p_value / 2
# Оценка результатов
if t_stat > 0 and p_value_one_sided < 0.05:
    print("Отвергаем нулевую гипотезу: новая рекламная кампания увеличивает продажи")
else:
    print("Не отвергаем нулевую гипотезу: недостаточно доказательств, что новая рекламная кампания увеличивает продажи")
```

```python
# Установка стиля для графиков
sns.set(style="whitegrid")

# Визуализация данных
plt.figure(figsize=(14, 7))
sns.histplot(sales_no_ad, kde=True, color='blue', label='Не видели новую рекламу', bins=20, stat="density", alpha=0.6)
sns.histplot(sales_with_ad, kde=True, color='red', label='Видели новую рекламу', bins=20, stat="density", alpha=0.6)
plt.axvline(np.mean(sales_no_ad), color='blue', linestyle='dashed', linewidth=2, label=f'Среднее без рекламы: {np.mean(sales_no_ad):.2f}')
plt.axvline(np.mean(sales_with_ad), color='red', linestyle='dashed', linewidth=2, label=f'Среднее с рекламой: {np.mean(sales_with_ad):.2f}')
plt.title(f'Независимый t-тест: Влияние рекламы на продажи\np-value = {p_value:.35f}')
plt.legend()
plt.show()
```

## Парный t-тест (ttest_rel)

Бизнес-задача  
Компания хочет проверить, увеличила ли новая методика обучения производительность сотрудников.   
Для этого сравнивают производительность до и после внедрения новой методики у одной и той же группы сотрудников

Какие будут гипотезы?

```python
display(create_spoiler(title, content_ttest_rel))
```

```python
# Данные для анализа
np.random.seed(42)
# Генерация данных для производительности до внедрения новой методики
# loc=70: Среднее значение производительности до внедрения
# scale=5: Стандартное отклонение производительности до внедрения
# size=100: Количество наблюдений до внедрения
performance_before = np.random.normal(loc=70, scale=5, size=100)
# Генерация данных для производительности после внедрения новой методики
# loc=75: Среднее значение производительности после внедрения
# scale=5: Стандартное отклонение производительности после внедрения
# size=100: Количество наблюдений после внедрения
performance_after = np.random.normal(loc=75, scale=5, size=100)
```

```python
# Выполнение парного t-теста для сравнения средних значений до и после внедрения методики
t_stat, p_value = st.ttest_rel(performance_before, performance_after, alternative='less') # коммент alternative, alternative='greater'

print('p-value', p_value)

# Оценка результатов
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: новая методика обучения увеличивает производительность")
else:
    print("Не отвергаем нулевую гипотезу: недостаточно доказательств, что новая методика обучения увеличивает производительность")

print(f'Среднее до: {np.mean(performance_before):.2f}')
print(f'Среднее после: {np.mean(performance_after):.2f}')
```

```python
# Визуализация данных
plt.figure(figsize=(14, 7))
sns.histplot(performance_before, kde=True, color='blue', label='Производительност до', bins=20, stat="density", alpha=0.6)
sns.histplot(performance_after, kde=True, color='red', label='Производительность после', bins=20, stat="density", alpha=0.6)
plt.axvline(np.mean(performance_before), color='blue', linestyle='dashed', linewidth=2, label=f'Среднее до: {np.mean(performance_before):.2f}')
plt.axvline(np.mean(performance_after), color='red', linestyle='dashed', linewidth=2, label=f'Среднее после: {np.mean(performance_after):.2f}')
plt.title(f'Парный t-тест: Влияние новой методики обучения на производительность\np-value = {p_value:.32f}')
plt.legend()
plt.show()
```

## Одновыборочный t-тест (ttest_1samp)

Бизнес-задача  
Компания хочет проверить, соответствует ли средний уровень удовлетворенности клиентов установленному стандарту 8 (из 10)  

Какие будут гипотезы?

```python
display(create_spoiler(title, content_ttest_1samp))
```

```python
# Данные для анализа
np.random.seed(42)
# Генерация данных для уровня удовлетворенности клиентов
# loc=8: Среднее значение уровня удовлетворенности клиентов
# scale=0.5: Стандартное отклонение уровня удовлетворенности клиентов
# size=100: Количество наблюдений (выборка из 100 клиентов)
satisfaction_scores = np.random.normal(loc=8, scale=0.5, size=100)
```

```python
satisfaction_scores
```

```python
# Выполнение одновыборочного t-теста для проверки гипотезы
# Проверка, отличается ли средний уровень удовлетворенности клиентов от 8
t_stat, p_value = st.ttest_1samp(satisfaction_scores, 8)

print('p-value', p_value)

# Оценка результатов
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: уровень удовлетворенности клиентов отличается от 8")
else:
    print("Не отвергаем нулевую гипотезу: недостаточно доказательств, что уровень удовлетворенности клиентов отличается от 8")

print(f'Средняя удовлетворенность: {np.mean(satisfaction_scores):.2f}')
```

```python
# Визуализация данных
plt.figure(figsize=(14, 7))
sns.histplot(satisfaction_scores, kde=True, color='green', label='Уровень удовтетворенности', bins=20, stat="density", alpha=0.6)
plt.axvline(np.mean(satisfaction_scores), color='green', linestyle='dashed', linewidth=2, label=f'Средняя удовлетворенность: {np.mean(satisfaction_scores):.2f}')
plt.axvline(8, color='black', linestyle='dashed', linewidth=2, label='Стандарт удовлетворенности: 8')
plt.title(f'Одновыборочный t-тест: Уровень удовлетворенности клиентов\np-value = {p_value:.5f}')
plt.legend()
plt.show()
```

## Резюме

- T-test для независимых выборок (ttest_ind)  
Сравнение средних значений двух независимых групп.  
Пример: Сравнение успеваемости двух разных классов.  

- Одновыборочный t-тест (ttest_1samp)  
Сравнение среднего значения выборки с известным значением.  
Пример: Проверка, соответствует ли средняя оценка в классе установленному стандарту.  

- Парный t-тест (ttest_rel)  
Сравнение средних значений двух зависимых выборок (например, до и после эксперимента).  
Сравнение успеваемости учеников до и после внедрения новой методики обучения.  

## О параметре alternative в t-тесте

На примере выполнения парного t-теста для сравнения средних значений до и после внедрения методики, можно использовать параметр alternative для указания типа гипотезы, которую вы проверяете

```python
import scipy.stats as st

# Выполнение парного t-теста для сравнения средних значений до и после внедрения методики
t_stat, p_value = st.ttest_rel(performance_before, performance_after, alternative='less')
```

Параметр alternative:
- 'two-sided' (по умолчанию, двусторонний): Проверяет, отличаются ли средние значения двух выборок (независимо от направления). Используется для двусторонней гипотезы.
- 'less' (односторонний тест): Проверяет, меньше ли среднее значение первой выборки по сравнению со второй.
- 'greater' (односторонний тест): Проверяет, больше ли среднее значение первой выборки по сравнению со второй.


В каком порядке ставить данные:
- Важно учитывать порядок данных при использовании параметров 'less' и 'greater', так как они определяют направление тестирования.
- Например, если вы хотите проверить, улучшилась ли успеваемость после внедрения методики, то данные performance_before должны идти первыми, а performance_after — вторыми

```python
t_stat, p_value = st.ttest_rel(performance_before, performance_after, alternative='greater')
```

Это проверяет гипотезу о том, что среднее значение успеваемости после методики больше, чем до методики

Зачем это нужно:
- Использование параметра alternative позволяет более точно формулировать и проверять гипотезы, соответствующие вашему исследовательскому вопросу.
- Односторонние тесты могут быть более мощными (имеют больше шансов обнаружить эффект), если у вас есть конкретное предположение о направлении эффекта.
- Двусторонние тесты обеспечивают общий подход к проверке различий, не предполагая направления, что делает их более универсальными и частоиспользуемыми
